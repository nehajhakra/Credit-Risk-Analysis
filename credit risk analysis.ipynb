```python
#Basic Data Visualizations
# Distribution of Numerical Columns
customers.hist(figsize=(10, 6), bins=20)
plt.suptitle("Distribution of Numerical Columns in Customers Dataset")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Count of Loan Approvals (Bar Chart)
sns.countplot(x="approval_status", data=loans)  
plt.title("Loan Approval Status Count")  
plt.xticks(rotation=45)  # Rotate labels if needed  
plt.show()
```

```python
#Loan Amount Distribution (Histogram)
plt.hist(loans["loan_amount"], bins=10, color="skyblue", edgecolor="black")  
plt.title("Loan Amount Distribution")  
plt.xlabel("Loan Amount")  
plt.ylabel("Frequency")  
plt.show()
```

```python
#Loan Type Distribution (Bar Chart)
sns.countplot(x="loan_type", data=loans, palette="viridis")  
plt.title("Loan Type Count")  
plt.xticks(rotation=45)  
plt.show()

```

```python
#Customer Age Distribution (Histogram)

plt.hist(customers["age"], bins=10, color="lightcoral", edgecolor="black")  
plt.title("Age Distribution of Customers")  
plt.xlabel("Age")  
plt.ylabel("Count")  
plt.show()
```

```python
#Simple Transaction Amount Distribution Plot
import matplotlib.pyplot as plt

plt.hist(transactions["amount"], bins=15, color="limegreen", edgecolor="black")  
plt.title("Transaction Amount Distribution")  
plt.xlabel("Transaction Amount")  
plt.ylabel("Frequency")  
plt.show()
```
